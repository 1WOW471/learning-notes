# 线性回归的梯度函数

```python
import numpy as np
import math, copy
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([1.2, 2.4])
```

## 成本函数

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost = 0
    for i in range(m):
        f_wb = w * x[i] + b
        cost = cost + (f_wb - y[i] )**2
    total_cost = 1 / (2 * m) * cost
    return total_cost
```

### 梯度下降求和

结合线性模型预测$f_{w,b}(x^{(i)})$:
$$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$
利用训练数据减小成本来找到合适的参数w，b，从而缩小预测$f_{w,b}(x^{(i)})$和target$y^{(i)}$之间的误差，
$$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2\tag{2}$$

在这里梯度下降描述为：
$$\begin{align*} \text{repeat}&\text{ until convergence:} \; \lbrace \newline
\;  w &= w -  \alpha \frac{\partial J(w,b)}{\partial w} \tag{3}  \; \newline
 b &= b -  \alpha \frac{\partial J(w,b)}{\partial b}  \newline \rbrace
\end{align*}$$
上式参数w，b同时进行更新
梯度的定义如下：
$$
\begin{align}
\frac{\partial J(w,b)}{\partial w}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})x^{(i)} \tag{4}\\
  \frac{\partial J(w,b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)}) \tag{5}\\
\end{align}
$$
这里同时意为一次计算完成之后w，b同时进行更新

###  实现梯度下降


```python
def compute_gradient(x, y, w, b):
    """
    计算线性回归的梯度
    :param x:数据，有m个例子
    :param y: 目标值
    :param w: 模型参数
    :param b: 模型参数
    :return:
        dj_dw:对w的偏导数
        dj_db:对b的偏导数
    """

    m = x.shape[0]
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i =f_wb - y[i]
        dj_dw += dj_dw_i
        dj_db += dj_db_i
    dj_dw = dj_dw / m
    dj_db = dj_db / m

    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function,gradient_function):
    """
    执行梯度下降来找到合适的w,b.运用学习率并通过迭代来更新w，b
    :param x:例子
    :param y:目标值
    :param w_in:最初的模型参数
    :param b_in:最初的模型参数
    :param alpha:学习率
    :param num_itersm:运行梯度下降的迭代次数
    :param cost_function:成本函数
    :param gradient_function:梯度函数
    :return:
        w:更新之后的w
        b:更新之后的b
        j_history:成本函数的临时值
        p_history:参数的临时值
    """

    m = x.shape[0]
    w = copy.deepcopy(w_in)
    j_history =[]
    p_history =[]
    b = b_in
    w = w_in

    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient(x, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if i<100000:      # prevent resource exhaustion
            j_history.append( cost_function(x, y, w , b))
            p_history.append([w,b])
        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters/10) == 0:
            print(f"Iteration {i:4}: Cost {j_history[-1]:0.2e} ",
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")

    return w, b, j_history, p_history
```

```python
w_init = 0
b_init = 0
iterations = 10000
tmp_alpha = 1.0e-2
w_final ,b_final , j_his, p_his= gradient_descent(x_train,y_train,w_init,b_init, tmp_alpha, iterations,compute_cost,compute_gradient)
print(f"{w_final:8.4f},{b_final:8.4f}")
```
