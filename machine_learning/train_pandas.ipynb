### 数据聚合与分析

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Age': [24, 27, 22, 32, 29, 25, 30],
    'City': ['NYC', 'LA', 'Chicago', 'NYC', 'LA', 'Chicago', 'NYC'],
    'Salary': [50000, 60000, 55000, 70000, 80000, 75000, 85000]
}

df = pd.DataFrame(data)
print(df)
```

```python
grouped = df.groupby('City')['Salary'].mean()
print(grouped)
```

```python
aggregated = df.groupby('City')['Salary'].agg(['sum', 'count', 'max' , 'min'])
print(aggregated)
```

```python
#自定义聚合函数
std_salary = df.groupby('City')['Salary'].agg(lambda x: x.std())
print(std_salary)
```

#### 透视表和交叉表

```python
pivot = df.pivot_table(values = 'Salary', index = 'City', columns = 'Age', aggfunc = 'mean')
print(pivot)
```

```python
#交叉表crosstab() 查看类别与数据之间的关系
cross_tab = pd.crosstab(df['City'], df['Age'])
print(cross_tab)
```

#### 数据可视化

```python
import matplotlib.pyplot as plt

data = {'Year': [2018, 2019, 2020, 2021, 2022],
        'Sales': [100, 150, 200, 250, 300]}
df = pd.DataFrame(data)
#折线图
plt.plot(df['Year'], df['Sales'],marker='o')
plt.title('Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
#柱状图
# 示例数据
data = {'Category': ['A', 'B', 'C', 'D'],
        'Value': [30, 45, 20, 55]}
df = pd.DataFrame(data)

# 绘制柱状图
plt.bar(df['Category'], df['Value'])
plt.title('Category Comparison')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
#直方图
import numpy as np

# 生成随机数据
data = np.random.randn(1000)

# 绘制直方图
plt.hist(data, bins=30, edgecolor='black')
plt.title('Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#散点图
# 示例数据
data = {'X': [1, 2, 3, 4, 5],
        'Y': [1, 4, 9, 16, 25]}
df = pd.DataFrame(data)

# 绘制散点图
plt.scatter(df['X'], df['Y'])
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#箱型图
# 示例数据
data = np.random.randn(100)

# 绘制箱型图
plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```

```python
#热力图
import seaborn as sns

# 示例数据
data = np.random.rand(10, 12)

# 绘制热力图
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```
