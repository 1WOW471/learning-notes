# 线性回归

### 房屋模型
以线性回归模型来估计房屋价格，x单位为平方米，y单位为千元

```python
import numpy as np

# x_train 输入变量
#y_ train 输出变量

x_train = np.array([100.0, 120.0])
y_train = np.array([3314.0, 3976.8])
print(f"x_train={x_train}, y_train={y_train}")
```

```python
#线性回归函数
#暂时假设w，b为标量（后续学习中进行w与b的选择的讨论）
def computer_modle_output(x,w,b):
    m = x_train.shape[0] #训练集案例个数
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
w = 100
b = 100
tmp_f_wb = computer_modle_output(x_train,w,b)
print(tmp_f_wb)
#换一组w,b尝试
tmp_f_wb = computer_modle_output(x_train,1000,500)
print(tmp_f_wb)
```

tmp_f_wb的输出与y_train 的目标值相差太多，如何选择一对w，b能使模型刚好契合呢？这是需要到成本函数，一种学习算法。
